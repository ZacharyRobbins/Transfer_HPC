```python
FuelsDF=pd.read_csv('GrassFuelBreakdown.csv')
def calfrac(FuelsDF):
    FuelsFraction=pd.DataFrame()
    #FuelsDF['Total']=FuelsDF['FuelClass5']+FuelsDF['FuelClass6']
    FuelsFraction['Frac1']=FuelsDF['FuelClass1']/FuelsDF['Total']
    FuelsFraction['Frac2']=FuelsDF['FuelClass2']/FuelsDF['Total']
    FuelsFraction['Frac3']=FuelsDF['FuelClass3']/FuelsDF['Total']
    #FuelsFraction['Frac4']=FuelsDF['FuelClass4']/FuelsDF['Total']
    FuelsFraction['Frac5']=FuelsDF['FuelClass5']/FuelsDF['Total']
    FuelsFraction['Frac6']=FuelsDF['FuelClass6']/FuelsDF['Total']
    return(FuelsFraction)
def calSAV(FuelsDF,Sav):
    FuelsFraction=pd.DataFrame()
    #FuelsDF['Total']=FuelsDF['FuelClass5']+FuelsDF['FuelClass6']
    FuelsFraction['Frac1']=FuelsDF['FuelClass1']/FuelsDF['Total']
    FuelsFraction['Frac2']=FuelsDF['FuelClass2']/FuelsDF['Total']
    FuelsFraction['Frac3']=FuelsDF['FuelClass3']/FuelsDF['Total']
    #FuelsFraction['Frac4']=FuelsDF['FuelClass4']/FuelsDF['Total']
    FuelsFraction['Frac5']=FuelsDF['FuelClass5']/FuelsDF['Total']
    FuelsFraction['Frac6']=FuelsDF['FuelClass6']/FuelsDF['Total']
    FuelsSAV=pd.DataFrame()
    FuelsSAV['Frac1']=FuelsFraction['Frac1']*Sav[0]
    FuelsSAV['Frac2']=FuelsFraction['Frac2']*Sav[1]
    FuelsSAV['Frac3']=FuelsFraction['Frac3']*Sav[2]
    #FuelsFraction['Frac4']=FuelsFraction['Frac4']*Sav[3]
    FuelsSAV['Frac5']=FuelsFraction['Frac5']*Sav[4]
    #FuelsSAV['Frac6']=FuelsFraction['Frac6']*Sav[5]
    print(FuelsSAV.head())
    return(FuelsSAV.sum(axis='columns'))
```
